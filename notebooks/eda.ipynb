# Exploratory Data Analysis
Quick look at class balance and sample images before training.

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

DATA_DIR = '../data/raw'
classes = sorted(os.listdir(DATA_DIR))
print(f'Found {len(classes)} classes:')
for c in classes:
    n = len(os.listdir(os.path.join(DATA_DIR, c)))
    print(f'  {c}: {n} images')
```

```python
# Bar chart of class distribution -- important for spotting imbalance
counts = [len(os.listdir(os.path.join(DATA_DIR, c))) for c in classes]
plt.figure(figsize=(10, 5))
plt.bar(classes, counts)
plt.xticks(rotation=75, ha='right')
plt.ylabel('Number of images')
plt.title('Class distribution')
plt.tight_layout()
plt.show()
```

```python
# Sample images from a few classes
fig, axes = plt.subplots(1, min(5, len(classes)), figsize=(15, 4))
for ax, c in zip(axes, classes[:5]):
    sample_file = os.listdir(os.path.join(DATA_DIR, c))[0]
    img = Image.open(os.path.join(DATA_DIR, c, sample_file))
    ax.imshow(img)
    ax.set_title(c, fontsize=9)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
